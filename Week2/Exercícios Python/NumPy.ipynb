```python
# Difícil
# 1
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
sepallength = np.genfromtxt(url, delimiter = ",", dtype = "float", usecols = [0])
def softmax(y):
    exps = [np.exp(i) for i in y]
    s_exps = sum(exps)
    return np.array([e / s_exps for e in exps])
softmax(sepallength)
```

```python
# 2
iris_2d = np.genfromtxt(url, delimiter = ",", dtype = "float", usecols = [0, 1, 2, 3])
lines = list()
for row in iris_2d:
    if row[3] > 1.5 and row[1] < 5.0:
        lines.append(row)
np.array(lines)
```
